Univariate Analysis

```python
def numerical_plot(col_name):
    plot.subplots(figsize=(15, 5))
    plot.subplot(121)
    sns.distplot(dataset[col_name], color='black')
    plot.title(f"Distribution of {col_name}")
    plot.subplot(122)
    ax = sns.boxplot(y=dataset[col_name], showfliers=False)
    ax.axhline(dataset[col_name].median(), color='r', linestyle='--')
    plot.title(f"Boxplot of {col_name}")
    plot.tight_layout(pad=3)
```

```python
for i in dataset.columns:
    numerical_plot(i)
```

```python
graph= sns.pairplot(dataset, vars=['GRE Score', 'TOEFL Score', 'University Rating', 'SOP', 'CGPA'], hue='Research', corner=True)
plot.show()
```

```python
def scatter_plot(df, x_col, y_col, title):
    sns.scatterplot(x=x_col, y=y_col, data=df)
    plot.title(title)
    plot.xlabel(x_col)
    plot.ylabel(y_col)
    plot.grid(alpha=0.3)
    plot.show()

def box_plot(dataset, x_col, y_col, title):
    sns.boxplot(x=x_col, y=y_col, data=dataset, showmeans=True)
    plot.title(title)
    plot.xlabel(x_col)
    plot.ylabel(y_col)
    plot.grid(alpha=0.3)
    plot.show()

def plot_admissions_data(dataset, columns):
    for col in columns:
        if col in ['GRE Score', 'TOEFL Score','CGPA']:
            scatter_plot(dataset, col, 'Chance of Admit', f'Admission Rates by {col}')
        elif col == 'University Rating':
            box_plot(dataset, col, 'Chance of Admit', f'Admission Rates by {col}')
            plot.plot(dataset.groupby(col)['Chance of Admit'].mean(), marker='o', linestyle='--')
            plot.title(f'Mean Admission Rates by {col}')
            plot.xlabel(col)
            plot.ylabel('Chance of Admit')
            plot.grid(alpha=0.3)
            plot.show()
        elif col in ['SOP', 'LOR', 'Research']:
            box_plot(dataset, col, 'Chance of Admit', f'Admission Rates by {col}')
```

# Model Building


Spliting data into Train and Test (Features and Labels)


```python
X = dataset.iloc[:,0:7]
y = dataset.iloc[:,7]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=150)
```

Data normalization is important in order to represent data in comparable scales

MIN - MAX Scaler its Better use always the Scaler methods to improve the Accuracy

```python
from sklearn.preprocessing import MinMaxScaler
xs=MinMaxScaler()
X_train[X_train.columns] = xs.fit_transform(X_train[X_train.columns])
X_test[X_test.columns] = xs.transform(X_test[X_test.columns])
```

## Classification Models & Furher Analysis

In order to make classification models, outcome of the classification is defined as 'status' which takes value of True for the applicants that have more than 83% chance. Third quartile, 83%, is chosen as threshold since median of the chance data, 72%, is pretty high.

```python
import numpy as np
cy_train=[1 if chance > 0.83 else 0 for chance in y_train]
cy_train=np.array(cy_train)

cy_test=[1 if chance > 0.83 else 0 for chance in y_test]
cy_test=np.array(cy_test)
```

Finding Linear Regression and printing the values of Class labels

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
preds = model.predict(X_test)
from sklearn.metrics import mean_squared_error,r2_score
mean_squared_error(y_test,preds)
r2_score(y_test,preds)
from scipy.special import expit
def ok(s):
    if s>=0.5:
        return "YES"
    else:
        return "NO"
admission = y.apply(ok)
admission.value_counts()
```

Fitting logistic regression model and
Printing accuracy score & confusion matrix

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train, cy_train)

from sklearn.metrics import accuracy_score
print('Logistic regression accuracy: {:.3f}'.format(accuracy_score(cy_test, lr.predict(X_test))))
print('--------------------------------------')
from sklearn.metrics import classification_report
print(classification_report(cy_test, lr.predict(X_test)))

cy = lr.predict(X_test)
from sklearn.metrics import confusion_matrix
import seaborn as sns
lr_confm = confusion_matrix(cy, cy_test)
sns.heatmap(lr_confm, annot=True, fmt='.2f',xticklabels = ["Admitted", "Rejected"] , yticklabels = ["Admitted", "Rejected"] )
plot.ylabel('Actual Class')
plot.xlabel('Predicted Class')
plot.title('Logistic Regression')
plot.show()
```

Fitting random forest model and Printing accuracy score & confusion matrix

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train, cy_train)

print('Random Forest Accuracy: {:.3f}'.format(accuracy_score(cy_test, rf.predict(X_test))))
print('--------------------------------------')
print(classification_report(cy_test, rf.predict(X_test)))

cy = rf.predict(X_test)
rf_confm = confusion_matrix(cy, cy_test)
sns.heatmap(rf_confm, annot=True, fmt='.2f',xticklabels = ["Admitted", "Rejected"] , yticklabels = ["Admitted", "Rejected"] )
plot.ylabel('True class')
plot.xlabel('Predicted class')
plot.title('Random Forest')
plot.show()
```

Feature Importance for Random Forest Model

In order to represent the most important features which influence whether an applicant is admitted in college, random forest classifier is used.

```python
f_imp=pd.Series(rf.feature_importances_,index=X_train.columns).sort_values(ascending=False)
print(f_imp)
```

```python
features = X.columns
importances = rf.feature_importances_
indices = np.argsort(importances)

plot.title('Feature Importances')
plot.barh(range(len(indices)), importances[indices], color='b', align='center')
plot.yticks(range(len(indices)), [features[i] for i in indices])
plot.xlabel('Relative Importance')
plot.show()
```

Fitting support vector machine model and Printing accuracy score & confusion matrix

```python
from sklearn.svm import SVC
svc = SVC()
svc.fit(X_train, cy_train)

print('Support vector machine accuracy: {:.3f}'.format(accuracy_score(cy_test, svc.predict(X_test))))
print('--------------------------------------')
print(classification_report(cy_test, svc.predict(X_test)))

cy = svc.predict(X_test)
svc_confm = confusion_matrix(cy, cy_test)
sns.heatmap(svc_confm, annot=True, fmt='.2f',xticklabels = ["Admitted", "Rejected"] , yticklabels = ["Admitted", "Rejected"] )
plot.ylabel('True class')
plot.xlabel('Predicted class')
plot.title('Support Vector Machine')
plot.show()
```
